```python
import os
import requests
import pandas as pd
import datetime

from bs4 import BeautifulSoup

url = 'https://www.thsrc.com.tw/TimeTable/Search'
# url = 'https://www.thsrc.com.tw/tw/TimeTable/SearchResult'
```

```python
after_one_week = datetime.datetime.now() + datetime.timedelta(weeks=1)
after_one_week_format = after_one_week.strftime('%Y/%m/%d')
print('The date after one week - {}'.format(after_one_week_format))
```

```python
form_data = {
    'SearchType': 'S',
    'Lang': 'TW',
    'StartStation': 'TaiPei',
    'EndStation': 'TaiNan',
    'OutWardSearchDate': after_one_week_format,
    'OutWardSearchTime': '13:30',
    'DiscountType': ''
}
form_data
```

```python
resp = requests.post(url, data=form_data)
resp.encoding = 'utf-8'
resp
```
